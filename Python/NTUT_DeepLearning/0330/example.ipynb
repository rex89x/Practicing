```python
import tensorflow as tf
import datetime, os

#下載dataset
(x_train,y_train), (x_test,y_test) = tf.keras.datasets.mnist.load_data()
#資料標準化
x_train, x_test = x_train / 255.0, x_test / 255.0
```

# model 1

```python
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(10, activation = 'softmax')
],
name = 'MNIST_model_1')
model1.summary()

model1.compile( optimizer = 'adam',
                loss = 'sparse_categorical_crossentropy',
                metrics = ['accuracy'])

log_dir = os.path.join('logs', 'model1', datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
model_cback = tf.keras.callbacks.TensorBoard(log_dir, histogram_freq=1)

history = model1.fit(x_train, y_train,
            batch_size = 32,
            epochs = 100,
            verbose = 1,
            validation_split = 0.3,
            callbacks = [model_cback]
            )
#查找history儲存的資訊有哪些
history.history.keys()

#
print(model1.evaluate(x_test, y_test))

predictions = model1.predict_classes(x_test)
print(predictions)
```

## model2 --減少模型大小

## model3 --加入權重正規化

```python
from tensorflow.keras import regularizers

model1 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(64, activation = 'relu',kernel_regularizer=regularizers.l2(0.0001)),
    tf.keras.layers.Dense(64, activation = 'relu',kernel_regularizer=regularizers.l2(0.0001)),
    tf.keras.layers.Dense(10, activation = 'softmax')
],
name = 'MNIST_model_1')
model1.summary()

model1.compile( optimizer = 'adam',
                loss = 'sparse_categorical_crossentropy',
                metrics = ['accuracy'])

log_dir = os.path.join('logs', 'model1', datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
model_cback = tf.keras.callbacks.TensorBoard(log_dir, histogram_freq=1)

history = model1.fit(x_train, y_train,
            batch_size = 32,
            epochs = 100,
            verbose = 1,
            validation_split = 0.3,
            callbacks = [model_cback]
            )
#查找history儲存的資訊有哪些
history.history.keys()

#
print(model1.evaluate(x_test, y_test))

predictions = model1.predict_classes(x_test)
print(predictions)
```

## model4 --加入dropout(捨棄比率設定為30%)

```python
from tensorflow.keras.layers import Dropout
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(64, activation = 'relu'),Dropout(0.3),
    tf.keras.layers.Dense(64, activation = 'relu'),Dropout(0.3),
    tf.keras.layers.Dense(10, activation = 'softmax')
],
name = 'MNIST_model_1')
model1.summary()

model1.compile( optimizer = 'adam',
                loss = 'sparse_categorical_crossentropy',
                metrics = ['accuracy'])

log_dir = os.path.join('logs', 'model1', datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
model_cback = tf.keras.callbacks.TensorBoard(log_dir, histogram_freq=1)

history = model1.fit(x_train, y_train,
            batch_size = 32,
            epochs = 100,
            verbose = 1,
            validation_split = 0.3,
            callbacks = [model_cback]
            )
#查找history儲存的資訊有哪些
history.history.keys()

#
print(model1.evaluate(x_test, y_test))

predictions = model1.predict_classes(x_test)
print(predictions)
```
