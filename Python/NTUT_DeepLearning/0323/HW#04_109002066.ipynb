```python
#matplotlib繪圖

#import套件
import matplotlib.pyplot as plt
import numpy as np
from random import randint

#汽車廠牌銷售量dataset
Benz = [3367,4120,5539,6020,6620]
BMW = [4000,3590,4423,4900,4590]
Lexus = [5200,4930,5350,6200,6930]

#汽車售出年度順序
year_seq = [2018,2019,2020,2021,2022]

#matplotlib圖表上下限與間隔
year_sort = [x for x in range(3000, 7001, 500)]
plt.plot(year_seq, Benz, '-*' , label='Benz')
plt.plot(year_seq, BMW, '-o' , label='BMW')
plt.plot(year_seq, Lexus, '-^', label='Lexus')

#宣告X,Y軸刻度
plt.xticks(year_seq)
plt.yticks(year_sort)

#matplotlib圖表內部設定
plt.legend(loc='upper left',bbox_to_anchor=(0,1))
plt.tight_layout(pad=2)
plt.title("Sales Report", fontsize=24)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Sales", fontsize=14)
plt.tick_params(axis='both', labelsize=12, color='red')

#存檔
plt.savefig('out20_13.jpg', bbox_inches='tight')

#顯示圖表
plt.show()
```

```python
''#csv檔案輸出實作
import pandas as pd
import numpy as np
course = ['Chinese', 'English', 'Math', 'Natural', 'Society']
Chinese=[14,12,13,10,13]
English=[13,14,11,10,15]
Nature=[15,10,13,10,15]
Math=[15,9,12,8,15]
Society=[12,11,14,9,14]
score=[Chinese,English,Math,Nature,Society]

member = [[score[i][j] for i in range(0,5)] for j in range(0,5)]

df = pd.DataFrame(member, columns=course, index=np.arange(1,6))
total = [df.iloc[i].sum() for i in range(0,5)]
df['Total'] = total

df = df.sort_values(by='Total', ascending=False)
level = np.arange(1,6)
df['Ranking'] = level
df = df.sort_index()

avg = df.mean()
df.loc['Average']=avg
df.iat[5,6]='NaN'
print(df)
df.to_csv("109002066.csv")

```
